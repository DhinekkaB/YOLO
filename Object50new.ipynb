```python
!pip install ultralytics
!pip install matplotlib pandas numpy

# Check GPU availability
!nvidia-smi

# Basic imports
import os
import yaml
import torch
from ultralytics import YOLO
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from IPython.display import display, Image
import time
import glob
from tqdm.notebook import tqdm

# Mount Google Drive if needed
from google.colab import drive
drive.mount('/content/drive')

# 2. Data Preparation and Exploration
# Set up dataset path - adjust as needed
dataset_path = '/content/dataset'  # Update this to your dataset path

print("Exploring dataset structure...")
print(f"Train images: {len(glob.glob(f'{dataset_path}/train/images/*'))}")
print(f"Train labels: {len(glob.glob(f'{dataset_path}/train/labels/*'))}")
print(f"Valid images: {len(glob.glob(f'{dataset_path}/valid/images/*'))}")
print(f"Valid labels: {len(glob.glob(f'{dataset_path}/valid/labels/*'))}")
print(f"Test images: {len(glob.glob(f'{dataset_path}/test/images/*'))}")
print(f"Test labels: {len(glob.glob(f'{dataset_path}/test/labels/*'))}")

# Examine a few label files to understand the format and classes
def explore_labels(label_path, num_samples=5):
    label_files = glob.glob(f'{label_path}/*.txt')
    print(f"\nExamining {num_samples} sample label files from {label_path}:")

    # Initialize class counter
    class_counts = {}

    # Process all label files to count classes
    for label_file in tqdm(label_files):
        with open(label_file, 'r') as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) >= 1:
                    class_id = int(parts[0])
                    if class_id in class_counts:
                        class_counts[class_id] += 1
                    else:
                        class_counts[class_id] = 1

    # Show sample content
    for i, label_file in enumerate(label_files[:num_samples]):
        print(f"\nSample {i+1}: {os.path.basename(label_file)}")
        with open(label_file, 'r') as f:
            content = f.read()
            print(content if content else "Empty file")

    return class_counts

# Analyze class distribution in training set
train_class_counts = explore_labels(f'{dataset_path}/train/labels')

print("\nClass distribution in training set:")
for class_id, count in sorted(train_class_counts.items()):
    print(f"Class {class_id}: {count} instances")

# Visualize class distribution
plt.figure(figsize=(15, 6))
plt.bar(train_class_counts.keys(), train_class_counts.values())
plt.title('Class Distribution in Training Dataset')
plt.xlabel('Class ID')
plt.ylabel('Count')
plt.xticks(list(train_class_counts.keys()))
plt.tight_layout()
plt.show()

# 3. Create YAML Configuration File
# Note: Since we're working with the YOLO format where class IDs start from 0,
# we need to map these IDs to class names

# List of class names in order (0-49 for 50 classes)
class_names = ['Bag', 'Battery', 'Chair', 'Cup-mug', 'Door', 'Eye-glass', 'Fridge', 'Handbag', 'Keyboard', 'Laptop', 'Phone', 'Plastic bottle', 'Ruler', 'Screwdriver', 'Spoon', 'Stapler', 'TV', 'Table', 'Umbrella', 'Wallet', 'Watch', 'alarm_clock', 'book', 'bowl', 'box', 'calendar', 'chocolate', 'coin', 'couch', 'fire-extinguisher', 'flowerPot', 'glass-tumbler', 'hammer', 'headphones', 'helmet', 'jars', 'key', 'knife', 'mouse', 'multimeter', 'pen', 'pendrive', 'plate', 'printer', 'pushbutton', 'router', 'scissor', 'shoe', 'speaker', 'webcam']

# Create a YAML file for dataset configuration
data_yaml = {
    'path': dataset_path,
    'train': f'{dataset_path}/train/images',
    'val': f'{dataset_path}/valid/images',  # Note: using 'valid' instead of 'val'
    'test': f'{dataset_path}/test/images',
    'nc': len(class_names),  # number of classes
    'names': class_names
}

# Save the YAML file
with open(f'{dataset_path}/data.yaml', 'w') as f:
    yaml.dump(data_yaml, f, sort_keys=False)

print(f"YAML file created at {dataset_path}/data.yaml")
with open(f'{dataset_path}/data.yaml', 'r') as f:
    print(f.read())

# 4. Model Selection and Training
# For computational efficiency on T4, let's start with YOLOv8n or YOLOv8s
# Download a pre-trained model
model = YOLO('yolov8n.pt')  # Nano model (smallest, fastest)
# Alternatives:
# model = YOLO('yolov8s.pt')  # Small model (balance between speed and accuracy)
# model = YOLO('yolov8m.pt')  # Medium model (better accuracy, slower)

print("Starting model training...")

# Training with optimized parameters for our goals - using only valid parameters
results = model.train(
    data=f'{dataset_path}/data.yaml',
    epochs=100,
    imgsz=640,
    batch=16,          # Adjust based on your GPU memory
    patience=15,       # Early stopping patience
    optimizer='Adam',  # Adam optimizer can help with convergence
    lr0=0.001,         # Initial learning rate
    lrf=0.01,          # Final learning rate as a fraction of initial lr
    momentum=0.937,
    weight_decay=0.0005,
    warmup_epochs=3,
    warmup_momentum=0.8,
    warmup_bias_lr=0.1,
    box=7.5,           # Box loss gain
    cls=0.5,           # Class loss gain (lower for multi-class)
    dfl=1.5,           # Distribution focal loss gain
    nbs=64,            # Nominal batch size for weight decay
    overlap_mask=True,
    close_mosaic=10,   # Close mosaic augmentation in final epochs for stability
    amp=True,          # Automatic mixed precision for faster training
    device=0,          # Use GPU
    verbose=True,      # Print verbose information

    # Data augmentation parameters for better generalization
    mosaic=1.0,        # Mosaic augmentation (0-1)
    mixup=0.1,         # Mixup augmentation (0-1)
    copy_paste=0.1,    # Copy-paste augmentation (0-1)
    degrees=0.0,       # Image rotation (+/- deg)
    translate=0.1,     # Image translation (+/- fraction)
    scale=0.5,         # Image scale (+/- gain)
    shear=0.0,         # Image shear (+/- deg)
    perspective=0.0,   # Image perspective (+/- fraction)
    flipud=0.0,        # Image flip up-down (probability)
    fliplr=0.5,        # Image flip left-right (probability)
    hsv_h=0.015,       # Image HSV-Hue augmentation (fraction)
    hsv_s=0.7,         # Image HSV-Saturation augmentation (fraction)
    hsv_v=0.4,         # Image HSV-Value augmentation (fraction)
)

# 5. Model Evaluation
print("\nModel training complete. Running validation...")
val_results = model.val()

# Print metrics
print("\nValidation Results:")
metrics = val_results.box
print(f"Precision: {metrics.mp:.4f}")  # mean precision
print(f"Recall: {metrics.mr:.4f}")     # mean recall
print(f"mAP50: {metrics.map50:.4f}")   # mAP at IoU 0.5
print(f"mAP50-95: {metrics.map:.4f}")  # mAP at IoU 0.5-0.95

# Per-class metrics
print("\nPer-class metrics:")
for i, cls_name in enumerate(class_names):
    if i < len(metrics.ap_class_index):
        idx = np.where(metrics.ap_class_index == i)[0]
        if len(idx) > 0:
            ap50 = metrics.ap50[idx[0]]
            print(f"Class {i} ({cls_name}): AP50 = {ap50:.4f}")
        else:
            print(f"Class {i} ({cls_name}): No detections")
    else:
        print(f"Class {i} ({cls_name}): No detections")

# 6. Test on Unseen Data
print("\nRunning inference on test data...")
test_results = model.predict(
    source=f'{dataset_path}/test/images',
    save=True,
    conf=0.25,  # Confidence threshold
    iou=0.45,   # IoU threshold for NMS
    max_det=300,
    save_conf=True,
    save_txt=True
)

# Visualize some predictions (up to 5 examples)
num_test_examples = min(5, len(test_results))
print(f"\nVisualizing {num_test_examples} test predictions...")

for i in range(num_test_examples):
    result = test_results[i]
    img = result.orig_img
    boxes = result.boxes

    plt.figure(figsize=(10, 10))
    plt.imshow(img)

    # Draw bounding boxes and labels
    for box in boxes:
        # Convert tensors to CPU and then to NumPy arrays
        x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
        cls = int(box.cls[0].cpu().numpy())
        conf = float(box.conf[0].cpu().numpy())
        cls_name = class_names[cls] if cls < len(class_names) else f"Unknown-{cls}"
        label = f"{cls_name} {conf:.2f}"

        plt.gca().add_patch(plt.Rectangle((x1, y1), x2-x1, y2-y1,
                                         fill=False, color='red', linewidth=2))
        plt.text(x1, y1, label, color='white', fontsize=12,
                backgroundcolor='red')

    plt.axis('off')
    plt.title(f'Test Image {i+1}')
    plt.show()

# 7. Model Optimization for Lower Latency
print("\nOptimizing model for deployment...")

# Export to different formats for deployment
# ONNX format for inference acceleration
onnx_path = model.export(format='onnx', dynamic=True, simplify=True)
print(f"Model exported to ONNX: {onnx_path}")

# TorchScript format
torchscript_path = model.export(format='torchscript')
print(f"Model exported to TorchScript: {torchscript_path}")

# TFLite format (for mobile deployment)
tflite_path = model.export(format='tflite')
print(f"Model exported to TFLite: {tflite_path}")

# 8. Test Model Inference Speed
print("\nMeasuring inference speed...")

# Test inference speed on the exported ONNX model
onnx_model = YOLO(onnx_path)

# Get a test image
test_images = glob.glob(f'{dataset_path}/test/images/*')
if test_images:
    test_img = test_images[0]

    # Warmup
    print("Warming up model...")
    for _ in range(10):
        _ = onnx_model(test_img)

    # Benchmark
    print("Benchmarking inference time...")
    times = []
    num_inferences = 50  # Reduced from 100 to save time

    for _ in tqdm(range(num_inferences)):
        start = time.time()
        _ = onnx_model(test_img)
        end = time.time()
        times.append(end - start)

    avg_time = sum(times) / len(times)
    print(f"Average inference time: {avg_time*1000:.2f} ms")
    print(f"FPS: {1/avg_time:.2f}")

    # Plot inference time distribution
    plt.figure(figsize=(10, 5))
    plt.hist(np.array(times) * 1000, bins=20)
    plt.title('Inference Time Distribution')
    plt.xlabel('Time (ms)')
    plt.ylabel('Count')
    plt.grid(True)
    plt.show()
else:
    print("No test images found for benchmarking")

# 9. Class-based Analysis
print("\nAnalyzing per-class performance...")

# Find classes with low precision (if available)
if hasattr(val_results.box, 'ap_class_index') and hasattr(val_results.box, 'ap50'):
    low_precision_classes = []
    precision_threshold = 0.7  # Adjust as needed

    for i, idx in enumerate(val_results.box.ap_class_index):
        ap50 = val_results.box.ap50[i]
        if ap50 < precision_threshold:
            class_name = class_names[idx] if idx < len(class_names) else f"Unknown-{idx}"
            low_precision_classes.append((class_name, ap50))

    print("Classes with low precision (AP50 < 0.7):")
    for cls, prec in low_precision_classes:
        print(f"{cls}: {prec:.4f}")

    if not low_precision_classes:
        print("All classes have good precision!")
else:
    print("Detailed per-class metrics not available")

# 10. Save Final Model for Deployment
print("\nPreparing final model for deployment...")

# Save the model to a specific location (e.g., Google Drive)
best_model_path = model.ckpt_path
save_path = '/content/drive/MyDrive/yolo_object_detection_model'
os.makedirs(save_path, exist_ok=True)

# Copy the best model
import shutil
shutil.copy(best_model_path, f"{save_path}/best_model.pt")
print(f"Best model saved to {save_path}/best_model.pt")
```
